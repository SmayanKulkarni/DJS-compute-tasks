```python
columns_checking = ['player_tags','loaned_from','nation_position','nation_jersey_number','defending_marking','gk_diving','player_traits','club_name','league_name','league_rank','release_clause_eur','team_position',
                    'team_jersey_number','joined','contract_valid_until']

def null_pecentage(df, columns):
    result={}
    for col in columns:
        if col in df.columns:
            null_percentage = (df[col].isnull().sum() / len(df)) * 100
            result[col] = round(null_percentage, 2)
        else:
            result[col] = "Column not found"
    return result

null_percentage = null_pecentage(df, columns_checking)

for column, percentage in null_percentage.items():
    if isinstance(percentage, float):
        print(f"{column}: {percentage}%")
    else:
        print(f"{column}: {percentage}")
    
```

```python
sns.barplot(data=null_percentage, x=null_percentage.values(), y=null_percentage.keys(), palette='viridis', hue=null_percentage.keys()).set(title="Percentage of NaN Values", xlabel="Percentage (%)", ylabel="Features")
plt.show()
```

```python
df['height_m']  = df['height_cm']/100
```

```python
df.drop('height_cm', axis=1,inplace=True)
```

```python
df['bmi'] = df['weight_kg']/(df['height_m']**2)
```

```python
df['bmi']
```

```python
def bodtype(bmi):
    if bmi < 18.5:
        return 'skinny'
    elif 18.5 <= bmi < 20:
        return 'lean'
    elif 20 <= bmi < 24.9:
        return 'muscular'
    elif 24.9 <= bmi < 25:
        return 'stocky'
    else:
        return 'obese'
```

```python
df['Body_Type'] = df['bmi'].apply(bodtype)
df.drop('body_type', axis=1, inplace=True)
```
