# Useful Seaborn Charts with Use Cases

This notebook covers some of the most useful Seaborn charts, along with explanations of when to use each chart type.

---

## 1. Bar Plot

**Use Case:** A bar plot displays the relationship between a categorical variable and a numeric variable. Itâ€™s useful for comparing average values across categories.

**Example:** Comparing average sales across different regions.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = {
    'category': ['Region A', 'Region B', 'Region C', 'Region D'],
    'value': [250, 300, 150, 400]
}

# Bar plot
sns.barplot(x='category', y='value', data=data)
plt.title("Average Sales by Region")
plt.show()
```

---

## 2. Count Plot

**Use Case:** A count plot shows the frequency of each category in a dataset. Itâ€™s ideal for understanding the distribution of a categorical variable.

**Example:** Counting the number of occurrences of different product types.


```python
# Example data
data = {
    'product': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type A']
}

# Count plot
sns.countplot(x='product', data=data)
plt.title("Count of Product Types")
plt.show()
```

---

## 3. Box Plot

**Use Case:** A box plot visualizes the distribution of a numeric variable through quartiles, medians, and outliers. Itâ€™s great for comparing distributions across multiple groups.

**Example:** Comparing test scores across different classes.


```python
# Example data
data = {
    'class': ['Class 1', 'Class 2', 'Class 1', 'Class 3', 'Class 2', 'Class 3'],
    'score': [78, 85, 74, 90, 82, 88]
}

# Box plot
sns.boxplot(x='class', y='score', data=data)
plt.title("Test Scores by Class")
plt.show()
```

---

## 4. Violin Plot

**Use Case:** A violin plot combines aspects of a box plot and a kernel density plot. Itâ€™s useful for visualizing both the distribution and probability density of numeric data across categories.

**Example:** Visualizing the distribution of salaries across different departments.


```python
# Example data
data = {
    'department': ['HR', 'IT', 'HR', 'Sales', 'IT', 'Sales'],
    'salary': [40000, 60000, 45000, 55000, 65000, 52000]
}

# Violin plot
sns.violinplot(x='department', y='salary', data=data)
plt.title("Salary Distribution by Department")
plt.show()
```

---

## 5. Scatter Plot

**Use Case:** A scatter plot shows the relationship between two numeric variables. Itâ€™s ideal for identifying correlations or clusters.

**Example:** Analyzing the relationship between age and income.


```python
# Example data
data = {
    'age': [22, 25, 30, 35, 40, 45],
    'income': [35000, 40000, 50000, 55000, 60000, 65000]
}

# Scatter plot
sns.scatterplot(x='age', y='income', data=data)
plt.title("Age vs. Income")
plt.show()
```

---

## 6. Line Plot

**Use Case:** A line plot displays trends over time or continuous data. Itâ€™s commonly used for time series analysis.

**Example:** Plotting sales trends over months.


```python
# Example data
data = {
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'sales': [100, 120, 150, 170, 200, 230]
}

# Line plot
sns.lineplot(x='month', y='sales', data=data)
plt.title("Sales Trend Over Time")
plt.show()
```

---

## 7. Heatmap

**Use Case:** A heatmap visualizes matrices or correlations using color gradients. Itâ€™s great for spotting patterns or anomalies.

**Example:** Correlation matrix of different financial indicators.


```python
import numpy as np

# Example data
data = np.random.rand(4, 4)
correlation_matrix = np.corrcoef(data)

# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

---

## 8. Pair Plot

**Use Case:** A pair plot shows pairwise relationships using scatter plots and histograms. Itâ€™s commonly used for exploratory data analysis.

**Example:** Analyzing relationships between features in the Iris dataset.


```python
from seaborn import load_dataset

# Load example dataset
iris = load_dataset('iris')

# Pair plot
sns.pairplot(iris)
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

---

## 9. Joint Plot

**Use Case:** A joint plot combines a scatter plot with histograms or density plots on the margins. It allows detailed examination of bivariate relationships.

**Example:** Examining the relationship between height and weight.


```python
# Example data
data = {
    'height': [150, 160, 165, 170, 175, 180],
    'weight': [55, 60, 65, 70, 75, 80]
}

# Joint plot
sns.jointplot(x='height', y='weight', data=data, kind='scatter')
plt.title("Height vs. Weight")
plt.show()
```

# More Useful Seaborn Charts

This section explores additional Seaborn charts and their use cases.

---

## 10. Swarm Plot

**Use Case:** A swarm plot is a categorical scatter plot that arranges data points along the x-axis while avoiding overlap. Itâ€™s ideal for showing the distribution of individual data points in relation to a categorical variable.

**Example:** Visualizing the distribution of test scores by gender.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
data = {
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'score': [85, 90, 78, 92, 88, 85]
}

# Swarm plot
sns.swarmplot(x='gender', y='score', data=data)
plt.title("Test Scores by Gender")
plt.show()
```

---

## 11. Strip Plot

**Use Case:** A strip plot is similar to a swarm plot but allows overlap of data points. Itâ€™s useful for showing the distribution of data across categories and can be combined with other plots for enhanced visualization.

**Example:** Visualizing the spread of customer ratings across product categories.


```python
# Example data
data = {
    'product': ['A', 'B', 'A', 'C', 'B', 'C'],
    'rating': [3, 4, 2, 5, 3, 4]
}

# Strip plot
sns.stripplot(x='product', y='rating', data=data, jitter=True)
plt.title("Customer Ratings by Product")
plt.show()
```

---

## 12. FacetGrid

**Use Case:** FacetGrid is used to plot multiple graphs based on the values of a categorical variable. Itâ€™s great for visualizing distributions or trends across subsets of data.

**Example:** Visualizing the distribution of total bills in a dataset based on gender and time.


```python
from seaborn import load_dataset

# Load example dataset
tips = load_dataset('tips')

# FacetGrid with histograms
g = sns.FacetGrid(tips, col="sex", hue="time", col_wrap=2)
g.map(plt.hist, "total_bill", bins=10, alpha=0.5)
g.add_legend()
plt.show()
```

---

## 13. KDE Plot

**Use Case:** Kernel Density Estimate (KDE) plots show the probability density of a continuous variable. They are useful for visualizing the distribution and smoothing out histograms.

**Example:** Plotting the distribution of customer ages.


```python
# Example data
ages = [20, 22, 25, 28, 30, 32, 35, 38, 40, 42, 45]

# KDE plot
sns.kdeplot(ages, shade=True)
plt.title("KDE Plot of Customer Ages")
plt.show()
```

---

## 14. Histogram

**Use Case:** A histogram plots the frequency distribution of a single numeric variable. Itâ€™s great for understanding how values are distributed across different ranges.

**Example:** Distribution of house prices in a neighborhood.


```python
# Example data
prices = [100000, 120000, 150000, 180000, 200000, 250000, 300000]

# Histogram plot
sns.histplot(prices, bins=5, kde=True)
plt.title("Distribution of House Prices")
plt.show()
```

---

## 15. Heatmap with Annotations

**Use Case:** A heatmap with annotations allows you to visualize matrix data (like correlations) while including text labels for better interpretation.

**Example:** A heatmap showing correlations between variables in a dataset.


```python
import numpy as np

# Example data
data = np.random.rand(4, 4)
correlation_matrix = np.corrcoef(data)

# Heatmap with annotations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

---

## 17. Pair Plot

**Use Case:** A pair plot visualizes pairwise relationships between multiple variables using scatter plots and histograms. Itâ€™s commonly used for exploratory data analysis, particularly in multivariate datasets.

**Example:** Exploring relationships between features in the Iris dataset.


```python
from seaborn import load_dataset

# Load the Iris dataset
iris = load_dataset('iris')

# Pair plot
sns.pairplot(iris, hue="species")
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

---

## 19. Cat Plot (Categorical Plot)

**Use Case:** A categorical plot (or catplot) is a flexible multi-plot function that can show different types of categorical plots such as bar plots, box plots, and more based on a specified kind. Itâ€™s ideal for comparing categorical variables across multiple subplots.

**Example:** Analyzing tips given by day of the week.


```python
# Load the tips dataset
tips = load_dataset('tips')

# Cat plot (box plot)
sns.catplot(x='day', y='tip', kind='box', data=tips)
plt.title("Box Plot of Tips by Day")
plt.show()
```
